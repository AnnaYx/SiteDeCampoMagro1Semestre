# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.




### Passo 1: Definir a Integral

A área $ A $ sob a curva $ y = x^5(2 - x) $ é dada pela integral:

$$
A = \int_{0}^{2} x^5(2 - x) \, dx
$$

### Passo 2: Expandir a Integral

Expandimos a expressão dentro da integral:

$$
A = \int_{0}^{2} (2x^5 - x^6) \, dx
$$

### Passo 3: Separar a Integral

Podemos separar a integral em duas partes:

$$
A = \int_{0}^{2} 2x^5 \, dx - \int_{0}^{2} x^6 \, dx
$$

### Passo 4: Integrar Cada Termo

Agora, integramos cada termo separadamente.

Para o primeiro termo:

$$
\int_{0}^{2} 2x^5 \, dx = 2 \int_{0}^{2} x^5 \, dx
$$

Sabemos que:

$$
\int x^5 \, dx = \frac{x^6}{6}
$$

Então:

$$
2 \int_{0}^{2} x^5 \, dx = 2 \left[ \frac{x^6}{6} \right]_{0}^{2} = 2 \left( \frac{2^6}{6} - \frac{0^6}{6} \right) = 2 \left( \frac{64}{6} \right) = \frac{128}{6}
$$

Para o segundo termo:

$$
\int_{0}^{2} x^6 \, dx = \left[ \frac{x^7}{7} \right]_{0}^{2} = \frac{2^7}{7} - \frac{0^7}{7} = \frac{128}{7}
$$

### Passo 5: Subtrair os Resultados

Subtraímos os resultados das duas integrais:

$$
A = \frac{128}{6} - \frac{128}{7}
$$

### Passo 6: Simplificar a Expressão

Para simplificar a expressão, encontramos um denominador comum:

$$
\frac{128}{6} = \frac{128 \cdot 7}{6 \cdot 7} = \frac{896}{42}
$$

$$
\frac{128}{7} = \frac{128 \cdot 6}{7 \cdot 6} = \frac{768}{42}
$$

Assim, a diferença é:

$$
A = \frac{896}{42} - \frac{768}{42} = \frac{128}{42}
$$

### Passo 7: Simplificar a Fração

Simplificamos a fração:

$$
\frac{128}{42} = \frac{64}{21} \approx 3.\overline{047619}
$$

Portanto, a área sob a curva, que representa a quantidade total de biomassa acumulada nos primeiros 2 anos, é aproximadamente $ 3.\overline{047619} $ metros.

# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate

# Definir a função
x = symbols('x')
F = x**5 * (2 - x)

# Calcular a área sob a curva usando integração simbólica
area = integrate(F, (x, 0, 2))

# Converter a área para um valor numérico
area_value = float(area.evalf())

# Gerar valores para o gráfico
x_vals = np.linspace(0, 2, 400)
F_vals = x_vals**5 * (2 - x_vals)

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_vals, F_vals, label=r'$F(x) = x^5 (2 - x)$', color='blue')
plt.fill_between(x_vals, F_vals, where=[(0 <= xi <= 2) for xi in x_vals],
                 color='skyblue', alpha=0.4, hatch='//', label=f'Área = {area_value:.4f}')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Gráfico de $F(x) = x^5 (2 - x)$ e a Área sob a Curva')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

print(f'A área sob a curva de f(x) entre x=0 e x=2 é aproximadamente {area.evalf()}')
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.



Dada a aceleração:

$$ a(t) = \cos(\pi t) $$

E a condição inicial da velocidade:

$$ v(0) = \frac{1}{2\pi} $$

Para encontrar a velocidade $ v(t) $ precisamos integrar a aceleração:

$$ v(t) = \int \cos(\pi t) \, dt $$

Usamos a substituição:

$$ u = \pi t $$
$$ du = \pi \, dt $$
$$ dt = \frac{du}{\pi} $$

Substituindo na integral:

$$ v(t) = \int \cos(u) \frac{du}{\pi} = \frac{1}{\pi} \int \cos(u) \, du $$

Integramos:

$$ v(t) = \frac{\sin(u)}{\pi} + C $$
$$ v(t) = \frac{\sin(\pi t)}{\pi} + C $$

Usamos a condição inicial $ v(0) = \frac{1}{2\pi}$) para encontrar $ C $

$$ v(0) = \frac{\sin(0)}{\pi} + C = 0 + C = \frac{1}{2\pi} $$

Portanto:

$$ C = \frac{1}{2\pi} $$

Logo, a velocidade é:

$$ v(t) = \frac{\sin(\pi t)}{\pi} + \frac{1}{2\pi} $$

Agora, para calcular a distância percorrida, integramos a velocidade de $ t = 0 $ a $ t = 1.5 $

$$ d = \int_{0}^{1.5} \left( \frac{\sin(\pi t)}{\pi} + \frac{1}{2\pi} \right) dt $$

Separando as integrais:

$$ d = \int_{0}^{1.5} \frac{\sin(\pi t)}{\pi} \, dt + \int_{0}^{1.5} \frac{1}{2\pi} \, dt $$

Para a primeira integral, usamos a substituição $ u = \pi t $

$$ \int_{0}^{1.5} \frac{\sin(\pi t)}{\pi} \, dt = \frac{1}{\pi} \int_{0}^{1.5} \sin(\pi t) \, dt $$
$$ = \frac{1}{\pi} \left[ -\frac{\cos(\pi t)}{\pi} \right]_{0}^{1.5} $$
$$ = -\frac{1}{\pi^2} \left[ \cos(\pi t) \right]_{0}^{1.5} $$
$$ = -\frac{1}{\pi^2} \left[ \cos(1.5\pi) - \cos(0) \right] $$
$$ = -\frac{1}{\pi^2} \left[ 0 - 1 \right] $$
$$ = \frac{1}{\pi^2} $$

Para a segunda integral:

$$ \int_{0}^{1.5} \frac{1}{2\pi} \, dt = \frac{1}{2\pi} \int_{0}^{1.5} 1 \, dt $$
$$ = \frac{1}{2\pi} \left[ t \right]_{0}^{1.5} $$
$$ = \frac{1.5}{2\pi} $$

Somando os dois resultados:

$$ d = \frac{1}{\pi^2} + \frac{1.5}{2\pi} $$

Considerando $\pi \approx 3.14$

$$ d \approx \frac{1}{(3.14)^2} + \frac{1.5}{2 \cdot 3.14} $$
$$ d \approx \frac{1}{9.8596} + \frac{1.5}{6.28} $$
$$ d \approx 0.1014 + 0.2389 $$
$$ d \approx 0.340 $$

Em $1.5$ segundos, a distância percorrida foi de aproximadamente $0.340$ metros.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from ipywidgets import interact, widgets
from IPython.display import display

# Definir as funções de velocidade e posição
def v(t):
    return (np.sin(np.pi * t) / np.pi) + (1 / (2 * np.pi))

def d(t):
    integral_v, _ = quad(v, 0, t)
    return integral_v

# Valores para o gráfico
t_vals = np.linspace(0, 1.5, 400)
v_vals = v(t_vals)
d_vals = [d(t) for t in t_vals]

# Plotar os gráficos
def plot_graphs():
    plt.figure(figsize=(12, 6))

    # Gráfico da Velocidade
    plt.subplot(2, 1, 1)
    plt.plot(t_vals, v_vals, label=r'$v(t) = \frac{\sin(\pi t)}{\pi} + \frac{1}{2\pi}$', color='blue')
    plt.title('Gráfico da Velocidade $v(t)$')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Velocidade $v(t)$')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()

    # Gráfico da Distância
    plt.subplot(2, 1, 2)
    plt.plot(t_vals, d_vals, label='Distância Percorrida $d(t)$', color='green')
    plt.fill_between(t_vals, d_vals, color='lightgreen', alpha=0.4)
    plt.title('Gráfico da Distância $d(t)$')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Distância $d(t)$')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()

    plt.tight_layout()
    plt.show()

# Adicionar interatividade com widgets
widget = widgets.Button(description="Atualizar Gráfico")

def on_button_clicked(b):
    plot_graphs()

widget.on_click(on_button_clicked)
display(widget)

plot_graphs()  # Mostrar o gráfico inicial
```

$$ \text{Média} = \frac{1}{b - a} \int_{a}^{b} f(x) \, dx $$

Nesse caso, $ a = 0 $ e $ b = 2 $ Substituindo na fórmula:

$$ \text{Média} = \frac{1}{2 - 0} \int_{0}^{2} (8x - 3 + 5e^{2-x}) \, dx $$

$$ \text{Média} = \frac{1}{2} \int_{0}^{2} (8x - 3 + 5e^{2-x}) \, dx $$

Para resolver a integral, precisamos separar as partes:

$$ \text{Média} = \frac{1}{2} \left( \int_{0}^{2} 8x \, dx - \int_{0}^{2} 3 \, dx + \int_{0}^{2} 5e^{2-x} \, dx \right) $$

Calculando cada integral separadamente:

$$ \int_{0}^{2} 8x \, dx = \left[ 4x^2 \right]_{0}^{2} = 4(2)^2 - 4(0)^2 = 16 $$

$$ \int_{0}^{2} 3 \, dx = \left[ 3x \right]_{0}^{2} = 3(2) - 3(0) = 6 $$

$$ \int_{0}^{2} 5e^{2-x} \, dx = \left[ -5e^{2-x} \right]_{0}^{2} = -5e^{0} + 5e^{2} = -5 + 5e^{2} $$

Substituindo os resultados de volta na fórmula da média:

$$ \text{Média} = \frac{1}{2} (16 - 6 - 5 + 5e^{2}) $$

$$ \text{Média} = \frac{1}{2} (5e^{2} + 5) $$

$$ \text{Média} = \frac{5}{2} (e^{2} + 1) $$

$$ \text{Média} \approx 26.26 $$

Portanto, a média da função $ f(x) $ no intervalo $ [0, 2] $ é aproximadamente $ 26.26 $.

b. Para encontrar a velocidade média do objeto no intervalo de tempo de $ t = 1 $ segundo até $ t = 4 $ segundos, usamos a fórmula da velocidade média:

$$ \text{Velocidade média} = \frac{\Delta x}{\Delta t} $$

Onde $ \Delta x $ é a mudança no deslocamento (ou posição) e $ \Delta t $ é a mudança no tempo.

Para calcular $ \Delta x $ no intervalo de tempo de $ t = 1 $ segundo até $ t = 4 $ segundos, substituímos $ t = 4 $ segundos e $ t = 1 $ segundo na função de deslocamento e calculamos a diferença:

$$ \Delta x = x(4) - x(1) $$

$$ \Delta x = (4^3 - 6(4)^2 + 9(4) + 1) - (1^3 - 6(1)^2 + 9(1) + 1) $$

$$ \Delta x = (64 - 96 + 36 + 1) - (1 - 6 + 9 + 1) $$

$$ \Delta x = 5 $$

A mudança no tempo é $ \Delta t = 4 - 1 = 3 $ segundos.

Agora podemos calcular a velocidade média:

$$ \text{Velocidade média} = \frac{\Delta x}{\Delta t} = \frac{5}{3} $$

Portanto, a velocidade média do objeto no intervalo de tempo de $ t = 1 $ segundo até $ t = 4 $ segundos é $ \frac{5}{3} $ metros por segundo.

$$ \int -6x^2 \, dx = -2x^3 + C_2 $$
$$ \int 2x \, dx = x^2 + C_3 $$
$$ \int -5 \, dx = -5x + C_4 $$

Juntando todos os termos, obtemos a família de antiderivadas:

$$ F(x) = x^4 - 2x^3 + x^2 - 5x + C $$

Aqui, $ C $ é uma constante de integração que pode assumir qualquer valor real.

b. Para a função $ f(x) = \frac{1}{x} + e^x $ para todo $ x \neq 0 $ podemos encontrar a antiderivada termo a termo.

$$ \int \frac{1}{x} \, dx = \ln|x| + C_1 $$
$$ \int e^x \, dx = e^x + C_2 $$

Juntando os termos, obtemos a família de antiderivadas:

$$ F(x) = \ln|x| + e^x + C $$

Aqui, $ C $ é uma constante de integração que pode assumir qualquer valor real. Note que a função $ \ln|x| $ é definida para todo $ x $ exceto $ x = 0 $ o que coincide com a restrição inicial da função $ f(x) $
